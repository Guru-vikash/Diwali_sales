```python
ax=sns.countplot(x='State' ,hue='Gender',data=df)

sns.set(rc={'figure.figsize':(20,10)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_state=df.groupby(['State'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)

#for clearly visible state names 
sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(x='State',y='Orders',data=sales_state)
```

```python
ax = sns.countplot(data=df ,x="Shaddi")

sns.set(rc={'figure.figsize':(7,5)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_state=df.groupby(['Shaddi','Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)

sns.set(rc={'figure.figsize':(6,5)})
sns.barplot(data=sales_state ,x='Shaddi' , y='Amount', hue='Gender' )
```
